```python
from flask import Flask, request
from flask_ngrok import run_with_ngrok
from IPython.display import display, HTML
import pandas as pd

app = Flask(__name__)
run_with_ngrok(app)  # Start ngrok when app is run


# Load the dataset
data = pd.read_csv('healthcare_dataset.csv')
data.dropna(inplace=True)


@app.route('/')
def index():
    form = """
    <h2>Enter Your Information</h2>
    <form method="post">
        <label for="name">Name:</label><br>
        <input type="text" id="name" name="name" required><br><br>

        <label for="gender">Gender:</label><br>
        <select id="gender" name="gender" required>
            <option value="Select">Select</option>
            <option value="Male">Male</option>
            <option value="Female">Female</option>
        </select><br><br>

        <label for="blood_group">Blood Group:</label><br>
        <select id="blood_group" name="blood_group" required>
            <option value="Select">Select</option>
            <option value="A+">A+</option>
            <option value="A-">A-</option>
            <option value="B+">B+</option>
            <option value="B-">B-</option>
            <option value="O+">O+</option>
            <option value="O-">O-</option>
            <option value="AB+">AB+</option>
            <option value="AB-">AB-</option>
            <!-- Add more options as needed -->
        </select><br><br>

        <input type="submit" value="Submit">
    </form>
    """
    return form

@app.route('/', methods=['POST'])
def submit():
    name = request.form['name']
    gender = request.form['gender']
    blood_group = request.form['blood_group']
    
     # Find medical condition based on user input
    condition = data[(data['Name'] == name) & (data['Gender'] == gender) & (data['Blood Group Type'] == blood_group)]['Medical Condition'].values
    
    if len(condition) > 0:
        result = f"<h2>Hello, {name}</h2><p>Based on your information, you are at most risk of: {condition[0]}</p>"
    else:
        result = f"<h2>Hello, {name}</h2><p>No medical condition found for the given information.</p>"
    
     # Store user data in CSV file
    with open('user_data.csv', 'a') as file:
        file.write(f"{name},{gender},{blood_group}\n")
    
    
    return result

if __name__ == '__main__':
    app.run()
```
